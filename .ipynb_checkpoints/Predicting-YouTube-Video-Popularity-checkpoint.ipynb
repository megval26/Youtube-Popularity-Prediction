## Feature Engineering

First, let's codify some terms. We've been using viral and popular interchangably. Let's define exactly what we mean. We'll define a video to be popular if it experiences high user interaction in the form of likes, dislikse or comments. Likewise, a video is non-popular if it does not experience high user interaction by those measures. The goal for this project is to determine what the features of popular videos are, and see if our model is good enough to predict the future success of videos. 

We will be using the `viral_meta.csv` dataset from CMU found [here](https://sites.google.com/site/cmuviralvideos/). We will also be using a list of random Youtube videos that we determined by inspection on whether it was popular or nonpopular. We determined first how popular the video was in relation to it's domain: for example a family video with 5000 views may be popular, while a Vevo video with 5000 views is definitely not popular. 


Knowing this, we will import this dataset into pandas by the following specifications

1) Drop all rows that have empty data (NaNs)

2) Add the following ratios that may be interesting to look at:
* likes to dislikes
* likes to views
* duration of video to number of views
* number of views to how long the video has been up
* number of dislikes to the number of views
* number of views to the numebr of subscribers

```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import svm
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier 
from sklearn.multiclass import OneVsRestClassifier
```

```python
def load_and_modify_data(file):
    """ Loads and transforms to data to include ratios and use popularity
    Inputs:
        csv: Takes a path to a csv file of the raw dataset
    Outputs:
        None
    """
    df = pd.read_csv(file,sep = '\t')
    df = df.dropna()    
    
    #One hot encoding of popularity
    viral_vids = pd.read_csv('viral_meta.csv')
    df['popularity'] = np.array([0 if vid.video_id in viral_vids['video_id'] else 0 for vid in viral_vids['video_id']])
    
    #Add ratios
    df['likes_to_dislikes'] = create_ratio(df,'num_likes','num_dislikes' )
    df['likes_to_views'] = create_ratio(df,'num_likes','view_count')
    df['duration_to_views'] = create_ratio(df, 'duration_in_seconds','view_count')
    df['views_to_daysActive'] = create_ratio(df, 'num_views','days_active')
    df['numReplies_to_numComments'] = create_ratio(df,'num_replies','num_comments')
    df['numComments_to_views'] = create_ratio(df,'num_comments','view_count')
    df['dislikes_to_views'] = create_ratio(df, 'num_dislikes','view_count')
    df['views_to_subscribers'] = create_ratio(df,'view_count','subscribers')
    
    
    #Split the dataset into 2 json files
    df_nonpop = df[df['Popularity'] == 0]
    df_pop = df[df['Popularity'] == 1]
    df_nonpop.to_csv('NonPopularVideoDataset.json')
    df_pop.to_csv('PopularVideoDataset.json')

    
#create_ratio takes in two column names in a dataframe and returns a ratio
#of the form: col1/col2. Values in col1 and col2 must be floats
def create_ratio(df, col1, col2):
    df[col1] = df[col1].astype(float)
    df[col2] = df[col2].astype(float)
    top = df[col1]
    bottom = df[col2]
    return top/bottom
    
    
df = load_and_modify_data('dataset.txt')

df_nonpop = pd.read_csv('NonPopularVids.csv')
df_pop = pd.read_csv('PopVids.csv')

print df_nonpop.head()
print df_pop.head()

```

## Training and Predicting

Because we have such a large list of features per given video, we'll be using a RandomForestClassifier, which fits multiple decision trees to various sub-samples of the dataset in order to increase the accuracy of the model. You can read more about the Random Forest Classifier [here](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html). Intuitively, a simple SVM implementation was our first thought as  Much of the following breakdown and structure of methods is similar to what we learned in the Text Classification Notebook as part of Homework 3. 

First, we'll set up our training and testing data to be usable for the our following predicition methods. Because we want to test our classifier on different data than the videos that we train it on, we'll split our popular and non-popular video sets. We'll set aside 40% of each of these sets for training, and use the rest for testing. 

```python
def setupTrainingAndTestingData(pop_df, nonPop_df):
    """ creates Training and Testing Data from dataframes
    Inputs:
        pop_df: pd.DataFrame: youtube videos that have classifcation: popular
        nonPop_df: pd.DataFrame: youtube videos that have classification: nonpopular
    Outputs:
        training_df: pd.DataFrame: combined popular and nonpopular videos used for training
        testing_df: pd.DataFrame: combined popular and nonpopular videos used for testing
    """
    # Get the slicing point of the popular and nonpopular datasets
    m = (len(pop_df)/100)*40
    n = (len(nonPop_df)/100)*40

    #Use 70% of the data for training, and the rest for testing
    pop_train = pop_df[0:m]
    pop_test = pop_df[m:]
    
    nonPop_train = nonPop_df[0:n]
    nonPop_test = nonPop_df[n:]
    
    #Combine all the training data(popular and nonpopular videos) together into one set
    training_df = pop_train.append(nonPop_train)
    
    #Combine all the training data(popular and nonpopular videos) together into one set
    testing_df = pop_test.append(nonPop_test)
    
    return training_df, testing_df

    
```

```python
train_df, test_df = setupTrainingAndTestingData(df_pop, df_nonpop)
print(len(df_pop))
print(len(df_nonpop))
print(len(train_df))
print(len(test_df))
```

We already have the class labels, popular or non-popular, so we'll just write a method that grabs this field and set it appropiately for the training or testing set.

```python
def create_labels(youtube_videos):
    """ creates the class labels from screen_name
    Inputs:
        youtube_videos: pd.DataFrame: videos from the training set
    Outputs:
        numpy.ndarray(int): dense binary numpy array of class labels determing popular(1) or nonpopular(0)
    """
    result = []
    popularityScore = youtube_videos['popularity']

    countPop = 0
    countNonPop = 0
    for score in popularityScore:
        if score == 1 :
            result.append(1)
            countPop = countPop + 1
        else:
            result.append(0)
            countNonPop = countNonPop + 1
    
    
    classLabels = np.array(result)
    return classLabels
```

```python
y_train = create_labels(train_df)
y_test = create_labels(test_df)
```

Here we set up our Random Forest Classifier. We tested various min_sample_leaf values to balance out the accuracy of our classifier with the efficiency of the runtime of the model. While we manually tested several of these values, another option to do would be to run a grid search on this parameter. 

```python
def run_random_forest(training_df):
    """ Creates the random forest classifier after training on the training set
    Inputs:
        training_df: pd.Dataframe: subset of the videos
    Outputs:
        sklearn.RandomForestClassifier: classifier of trained data
    """
    
    X_train = training_df.drop('video_id', 1).as_matrix()

    y_train = create_labels(training_df)

    clf = RandomForestClassifier(min_samples_leaf=75)
    return clf.fit(X_train, y_train)
```

```python
classifier = run_random_forest(train_df)
print classifier
```

## Analysis

Having a classifier isn't enough! The following method simply sets up an easy way to test the classifier against a testing dataset.

```python
def evaluate_classifier(classifier, testing_df, y_testing):
    """ evaluates a classifier based on a supplied validation data
    Inputs:
        classifier: sklearn.svm.classes.SVC: classifer to evaluate
        X_train: scipy.sparse.csr.csr_matrix: sparse matrix of features
        y_train: numpy.ndarray(int): dense binary vector of class labels
    Outputs:
        double: accuracy of classifier on the validation data
    """
    X_test = testing_df.drop('video_id', 1).as_matrix()
    return classifier.score(X_test, y_testing)
    


accuracy = evaluate_classifier(classifier, test_df, y_test)
print accuracy 
```

Finally, we set up a method that combines this entire process into one easy flow given a set of popular or nonpopular videos.

```python
def classify_videos(classifier, pop_df, nonPop_df):
    """ predicts class labels for raw tweet text
    Inputs:
        tfidf: sklearn.feature_extraction.text.TfidfVectorizer: the TfidfVectorizer object used on training data
        classifier: sklearn.svm.classes.SVC: classifier learnt
        unlabeled_tweets: pd.DataFrame: tweets read from tweets_test.csv
    Outputs:
        numpy.ndarray(int): dense binary vector of class labels for unlabeled tweets
    """
    trainData, testData = setupTrainingAndTestingData(pop_df, nonPop_df)
    matrix = testData.drop('video_id', 1).as_matrix()
    prediction = classifier.predict(matrix)
    return prediction

# AUTOLAB_IGNORE_START
classifier = run_random_forest(train_df)
y_pred = classify_videos(classifier, df_pop, df_nonpop)
# AUTOLAB_IGNORE_STOP
```
